## **KNN**

---

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X, y = df.drop("wine_class", axis="columns"), df["wine_class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

scaler = MinMaxScaler()
scaler_X_train = scaler.fit_transform(X_train)
scaler_X_test = scaler.transform(X_test)

model = KNeighborsClassifier(n_neighbors= 1)
model.fit(scaler_X_train, y_train)

y_pred = model.predict(scaler_X_test)

print(f"{y_pred=}")
print(f"{y_test.values=}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
```

```python
print(classification_report(y_test, y_pred))
```

## **Elbow plot to choose K**

NOTE: 
- important to divide so that we also have validation data (train|val|test_split) and do the elbow plot on  validation data.

- then train on both validation and training data.

- evaluate on test data.

---

```python
from sklearn.metrics import accuracy_score

error = 1-accuracy_score(y_test, y_pred)
print(f"{error=}")
error_list = []

for k in range(1, 50):
    model = KNeighborsClassifier(k)
    model.fit(scaler_X_train, y_train)
    y_pred = model.predict(scaler_X_test)
    error_list.append(1 - accuracy_score(y_test, y_pred))

plt.plot(range(1, 50), error_list, ".-")
plt.title("Elbow plot")
plt.xlabel("K Neighbors")
```

```python
model = KNeighborsClassifier(9)
model.fit(scaler_X_train, y_train)
y_pred = model.predict(scaler_X_test)

print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
```
